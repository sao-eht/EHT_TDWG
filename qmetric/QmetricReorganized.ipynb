```python
import numpy as np
import matplotlib.pyplot as plt
import qmetric
```

### Test 0 - pure noise, uniform sampling

```python
#FAKE DATA
#time in arbitrary units
t0 = np.arange(0,5000,5)
N = len(t0)

#variability-pure noise
sig = 1.
obs = np.random.normal(0.,sig, N)

#thermal error
obs_sig = np.asarray([sig]*N)

#save as file
data = np.asarray([t0, obs, obs_sig])
np.savetxt('fake_data0',np.transpose(data))

#plot the data
plt.figure(figsize=(8,5))
plt.errorbar(t0,obs,obs_sig)
plt.ylabel('signal')
plt.xlabel('time')
plt.show()

#calculate qmetric
q,dq = qmetric.qmetric('fake_data0')
print 'Q-value: ', q
print 'Q-value error: ', dq
```

### Test 1 - sinusoidal variability + pure noise, uniform sampling

```python
#FAKE DATA
#time
t0 = np.arange(0,5000,5)
N = len(t0)
snr = 5.

#variability-pure noise
sig = 1./snr
obs = np.sin(t0/200.)+ np.random.normal(0.,sig, N)

#thermal error
obs_sig = np.asarray([sig]*N)

#save as file
data = np.asarray([t0, obs, obs_sig])
np.savetxt('fake_data0',np.transpose(data))

#qmetric.inflate_noise('fake_data0', 10.)
#plot the data
plt.figure(figsize=(8,5))
plt.errorbar(t0,obs,obs_sig)
plt.ylabel('signal')
plt.xlabel('time')
plt.show()

#calculate qmetric
q,dq = qmetric.qmetric('fake_data0')
print 'Q-value: ', q
print 'Q-value error: ', dq
```

### Test 2 - sinusoidal variability + pure noise, uniform sampling, Q dependence on SNR

```python
SNRL = [0.5,1.,2.,3.,5.,10.] #list of snr to test
NumTests = 1000 #how many trials

Qmatrix = np.zeros((len(SNRL),NumTests))
Qerr_matrix = np.zeros((len(SNRL),NumTests))
for couSNR in range(len(SNRL)):
    for couMC in range(NumTests):
        
        #FAKE DATA
        #time
        t0 = np.arange(0,5000,50)
        N = len(t0)

        #variability-pure noise
        sig = 1./SNRL[couSNR]
        obs = 1.*np.sin(t0/200.)+ np.random.normal(0.,sig, N)

        #thermal error
        obs_sig = np.asarray([sig]*N)

        #save as file
        data = np.asarray([t0, obs, obs_sig])
        np.savetxt('fake_data0',np.transpose(data))

        #calculate qmetric
        q,dq = qmetric.qmetric('fake_data0')
        Qmatrix[couSNR,couMC] = q
        Qerr_matrix[couSNR,couMC] = dq
        
Q_by_snr = np.mean(Qmatrix,1)
dQ_by_snr = np.mean(Qerr_matrix,1)

plt.figure(figsize=(8,5))
plt.errorbar(SNRL,Q_by_snr,dQ_by_snr,fmt='o-',capsize=5)
plt.xlabel('SNR')
plt.ylabel('Q')
plt.show()
```

### Commment: even in the most idealized example we see limitations to the Q-metric application below SNR = 3

```python
plt.figure(figsize=(8,5))
plt.errorbar(SNRL,Q_by_snr,dQ_by_snr,fmt='o-',capsize=5)
plt.xlabel('SNR')
plt.ylabel('Q')
plt.show()
```

### Test 3 - randomized sampling density, random varying snr

```python
#FAKE DATA
#time
How_variable_snrL = [0.01,0.1, 0.5, 1.]
t0 = np.arange(0,5000,50)
N = len(t0)
How_nonuniform = 30.
t_nonuni = t0 + How_nonuniform*np.round(np.abs(np.random.normal(0.,1., N)))

#variability-pure noise
SNR0 = 3.
sig0 = 1/SNR0
How_variable_snr = 1.5
obs_sig = [np.abs(np.random.normal(sig0,np.abs(How_variable_snr*sig0))) for x in range(N)]
obs_noise = [np.random.normal(0.,obs_sig[x]) for x in range(N)]
obs = 1.*np.sin(t_nonuni/200.)+ np.asarray(obs_noise)

data = np.asarray([t_nonuni, obs, obs_sig])
np.savetxt('fake_data0',np.transpose(data))
plt.figure(figsize=(8,6))
plt.errorbar(t_nonuni,obs,obs_sig)
plt.show()
diftimeL = 0.
segtimeL = 8000.
q,dq = qmetric.qmetric('fake_data0', detrend_deg = 0., segtime = segtimeL, diftime=diftimeL)
q, dq
```

### Comment: large variation in thermal sigma(SNR variation) will further inhibit our abilities to detect variability

```python
SNRL = [0.01,0.1,0.5,1.,2.,4.,6.] #list of variabilities of variability
NumTests = 1000

Qmatrix = np.zeros((len(SNRL),NumTests))
Qerr_matrix = np.zeros((len(SNRL),NumTests))
for couSNR in range(len(SNRL)):
    for couMC in range(NumTests):
        
        #FAKE DATA
        #time
        t0 = np.arange(0,5000,50)
        N = len(t0)
        How_nonuniform = 30.
        t_nonuni = t0 + How_nonuniform*np.round(np.abs(np.random.normal(0.,1., N)))

        #variability-pure noise
        SNR0 = 3.
        sig0 = 1/SNR0
        How_variable_snr = SNRL[couSNR]
        obs_sig = [np.abs(np.random.normal(sig0,np.abs(How_variable_snr*sig0))) for x in range(N)]
        obs_noise = [np.random.normal(0.,obs_sig[x]) for x in range(N)]
        obs = 1.*np.sin(t_nonuni/200.)+ np.asarray(obs_noise)

        #save as file
        data = np.asarray([t_nonuni, obs, obs_sig])
        np.savetxt('fake_data0',np.transpose(data))

        #calculate qmetric
        q,dq = qmetric.qmetric('fake_data0')
        Qmatrix[couSNR,couMC] = q
        Qerr_matrix[couSNR,couMC] = dq
        
Q_by_snr = np.mean(Qmatrix,1)
dQ_by_snr = np.mean(Qerr_matrix,1)
```

```python
plt.figure(figsize=(8,5))
plt.errorbar(SNRL,Q_by_snr,dQ_by_snr,fmt='o-',capsize=5)
plt.xlabel('SNR variation')
plt.ylabel('Q')
plt.show()
```
